```python
df_1 = df.drop(columns=['month', 'airport'])
df_1
```

```python
#count the numbers of each status
df_2 = df_1.groupby(['carrier_name', 'year'])['Total time of delayed flight due to carrier', 'Total time of delayed flight due to weather', 'Total time of delayed flight due to national aviation system', 'Total time of delayed flight due to security', 'Total number of time delayed flight as a result of a former flight on the same airplane being late', 'Total time of delayed flight'].sum()
df_2
```

```python
# Group the data by year and airport
grouped_data = df_2.groupby(['carrier_name', 'year'])




# Calculate the total number of flights and delayed flights for each year and airport
total_flights = grouped_data.size()
delayed_flights = grouped_data['Total time of delayed flight'].sum()





# Calculate the percentage of delayed flights for each year and airport
delayed_percentage = (delayed_flights // total_flights) * 100


# Create a new DataFrame to display the delayed percentage for each year and airport
delayed_df = pd.DataFrame({'Total Flights': total_flights, 
                           'Delayed Flights': delayed_flights, 
                           'Delayed Percentage': delayed_percentage})

print(delayed_df)
```

```python
import pymongo

# Reset the index of the DataFrame
delayed_df = delayed_df.reset_index()

# Pivot the DataFrame to display delayed percentage by year and airport
delayed_df_pivot = delayed_df.pivot(index='year', columns='carrier_name', values='Delayed Percentage')

# Plot the bar graph
ax = delayed_df_pivot.plot(kind='bar', figsize=(10, 6))

# Set the title and labels for the graph
ax.set_title('Flight Delayed Percentage by Year and Carrier Name')
ax.set_xlabel('Year')
ax.set_ylabel('Delayed Percentage')

# Show the plot
plt.show()


#In this example, we first load the flight data into a pandas DataFrame and filter the data to include only the relevant columns. We then group the data by year and airport and calculate the total number of flights and delayed flights for each year and airport. We calculate the percentage of delayed flights for each year and airport and create a new DataFrame to display the delayed percentage for each year and airport. We reset the index of the DataFrame and pivot it to display delayed percentage by year and airport. Finally, we plot the bar graph using the plot() method on the pivoted DataFrame and set the title and labels for the graph. The resulting output will show a bar graph of the flight delayed percentage by year and airport.
```

```python

# Group the DataFrame by Airport
grouped_df = delayed_df.groupby('carrier_name')


# Plot the line graph
fig, ax = plt.subplots(figsize=(10, 6))


# Set the title and labels for the graph
ax.set_title('Flight Delayed Percentage by Year and Carrier Name')
ax.set_xlabel('Year')
ax.set_ylabel('Delayed Percentage')
ax.legend(loc=9)

# Show the plot
plt.show()

```

```python
import plotly.express as px




# Pivot the DataFrame to display delayed percentage by year and airport
delayed_df_pivot = delayed_df.pivot(index='carrier_name', columns='year', values='Delayed Percentage')

# Plot the map graph
fig = px.choropleth(delayed_df_pivot, locations=delayed_df_pivot.index, 
                    locationmode="country names", scope='world', animation_frame='year',
                    color_continuous_scale=px.colors.sequential.Plasma, color=delayed_df_pivot.columns)

# Set the title and labels for the graph
fig.update_layout(title='Flight Delayed Percentage by Year and Airport',
                  xaxis_title='year',
                  yaxis_title='carrier_name')

# Show the plot
fig.show()
```
